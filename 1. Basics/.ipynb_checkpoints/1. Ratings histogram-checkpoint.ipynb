# Ratings Histogram


In this notebook, we use PySpark to get the ratings histogram of the Movielens data set. 

First, we call the following libraries and tell the computer that we are going to run our script on our local system.

```python
from pyspark import SparkConf, SparkContext

# This library will only be used for plotting our results
import matplotlib.pyplot as plt

conf = SparkConf().setMaster('local').setAppName('ratings_histogram')
sc = SparkContext(conf = conf)
```

We load our data file into a Resilient Distributed Dataset (RDD) object.

```python
lines = sc.textFile('../datasets/ml-100k/u.data')
```

We extract the ratings column form the data and create a new RDD, named ratings, that only contain the rating values.

```python
ratings = lines.map(lambda x: x.split()[2])
```

We perform an action, on the RDD that contains the ratings, that counts how many times each rating occurs.

```python
frequencies = ratings.countByValue()
frequencies = dict(sorted(frequencies.items()))
```

Finally, let's plot the information that we just got from our data set.

```python
rate_values = list(frequencies.keys())
frequ_values = list(frequencies.values())
plt.bar(rate_values, frequ_values)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
